#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import matplotlib.pyplot as plt
subconjunto_sur['price_norm'] = subconjunto_sur.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
alcorcon = subconjunto_sur[subconjunto_sur['level5'] == 'Alcorcon']['price_norm']
fuenlabrada = subconjunto_sur[subconjunto_sur['level5'] == 'Fuenlabrada']['price_norm']
getafe = subconjunto_sur[subconjunto_sur['level5'] == 'Getafe']['price_norm']
leganes = subconjunto_sur[subconjunto_sur['level5'] == 'Leganes']['price_norm']

plt.figure(figsize=(10, 6))
plt.hist(alcorcon, bins=30, alpha=0.5, label='Alcorcon', color='blue', edgecolor='black')
plt.hist(fuenlabrada, bins=30, alpha=0.5, label='Fuenlabrada', color='green', edgecolor='black')
plt.hist(getafe, bins=30, alpha=0.5, label='Getafe', color='red', edgecolor='black')
plt.hist(leganes, bins=30, alpha=0.5, label='Leganes', color='purple', edgecolor='black')

plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.title('Distribución Normalizada de Precios por Población en el Cinturón Sur de Madrid')
plt.legend()
plt.show()
```

Al observar el gráfico de distribuciones normalizadas, notamos cómo los precios varían en cada población del Cinturón Sur de Madrid. Las diferencias en el ancho y altura de los picos reflejan variaciones en la concentración de precios, indicando que algunas poblaciones tienen mayor dispersión de precios en comparación con otras.
